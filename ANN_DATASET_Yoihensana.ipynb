```python
# Q.1. Dataset generation
import numpy as nup
import pandas as pan
# Initializing the empty lists to store the inputs and outputs
A = []
B = []
C = []
F = []

# Defining the ranges for the inputs (for Styblinski-Tang function, n=3) 
# The input data ranges from [-5,5]
# Taking step value as 0.7
step = 0.7

# Generating the dataset
for a in nup.arange(-5,5.1,step):
    for b in nup.arange(-5,5.1,step):
        for c in nup.arange(-5,5.1,step):
            f = 0.5*((a**4 - 16*a**2 + 5*a) + (b**4 - 16*b**2 + 5*b) + (c**4 - 16*c**2 + 5*c))
            A .append(round(a,1))   ## rounding off A to 1 decimal place
            B .append(round(b,1))   ## rounding off B to 1 decimal place
            C .append(round(c,1))   ## rounding off C to 1 decimal place
            F .append(f)
            
# Displaying the generated dataset
print (len(A))
for i in range(len(A)):
    print (f"a={A[i]}, b={B[i]}, c={C[i]} : f={F[i]}")

# Creating a dataframe from the lists
data = {'a': A, 'b': B, 'c': C, 'f':F}
df = pan.DataFrame(data)

# Specifying the file path where the Excel file will be saved
file_path = 'ANN_Dataset.xlsx'

# Saving the Data-Frame to an Excel file
df.to_excel(file_path)

```
